### Step 4: Load images

```python
from PIL import Image # using Pillow for image manipulation and handling -mt
from numpy import array
```

```python
# modified from stackoverflow.com/questions/36774431/how-to-load-images-from-a-directory-on-the-computer-in-python -mt

# return array of images

def loadImages(path):
    
    imagesList = os.listdir(path)
    loadedImages = []
    for image in imagesList:
        img = Image.open(path + image)
        loadedImages.append(img)

    return loadedImages
```

```python
cat_path = "/Users/minhgeneralassembly/Downloads/kaggle_catsanddogs/mini_train_cats/"
dog_path = "/Users/minhgeneralassembly/Downloads/kaggle_catsanddogs/mini_train_dogs/"
```

```python
# your images in an array
cat_imgs = loadImages(cat_path)
dog_imgs = loadImages(dog_path)
```

### Step 4.1: Crop/resize images

```python
# resize image, center crop using smaller dimension -mt
# modified from https://stackoverflow.com/questions/16646183/crop-an-image-in-the-centre-using-pil -mt

def cropResizeImages(img):
    width, height = img.size

    if width > height:
        left = (width - height)/2
        top = (height - height)/2
        right = (width + height)/2
        bottom = (height + height)/2
        
        img = img.crop((left, top, right, bottom))
        img = img.resize([250,250])
    
    elif height > width:
        left = (width - width)/2
        top = (height - width)/2
        right = (width + width)/2
        bottom = (height + width)/2
        
        img = img.crop((left, top, right, bottom))
        img = img.resize([250,250])
        
    return img
```

```python
## ask nico for a more efficient way to do this (instead of making a new list, replacing the old list (cat_img)?)
cat_img_cropped = []
for cat_img in cat_imgs:
    cat_img_cropped.append(cropResizeImages(cat_img))
```

### Step 4: Load image data form MNIST

```python
from keras.datasets import mnist

# Load pre-shuffled MNIST data into train and test sets
(X_train, y_train), (X_test, y_test) = mnist.load_data()
```

```python
print X_train.shape
```

```python
from matplotlib import pyplot as plt
%matplotlib inline
# ^ show stuff inline
plt.imshow(X_train[0], cmap=plt.cm.binary) #"cmap=plt.cm.binary" used to make image b/w
```

### Step 5: Preprocess input data for Keras

```python
# old shape (currently # of samples, height, width)
print X_train.shape
```

```python
# explicitly declaring the depth dimension
# need shape to look like # of samples, depth, height, width
X_train = X_train.reshape(X_train.shape[0], 1, 28, 28)
X_test = X_test.reshape(X_test.shape[0], 1, 28, 28)
```

```python
# new shape
print X_train.shape
```

```python
# Convert data type
X_train = X_train.astype('float32')
X_test = X_test.astype('float32')
```

```python
# normalize values
X_train /= 255
X_test /= 255
```

### Step 6: Preprocess class labels for Keras

```python
# taking a look at the shape of our class label data
print y_train.shape
```

```python
# expecting 10 classes (one for each digit)
```

```python
# looking at labels for first 10 training samples
print y_train[:10]
```

```python
# preprocess class labels
# Convert 1-dimensional class arrays to 10-dimensional class matrices
Y_train = np_utils.to_categorical(y_train, 10)
Y_test = np_utils.to_categorical(y_test, 10)
```

```python
# resulting shape
print Y_train.shape
```

### Step 7: Define model architecture

```python
# using a pre-defined / proven architecture for this
# declare sequential model
model = Sequential()
```

```python
# declare CNN input layer
# added dim_ordering='th' since i'm using tensorflow and it has different ways of ordering dimensions
# model.add(Convolution2D(32, (3, 3), activation='relu', input_shape=(1,28,28), dim_ordering='th'))
model.add(Convolution2D(32, (3, 3), activation='relu', input_shape=(1,28,28), data_format="channels_first"))
```

```python
print model.output_shape
```

```python
# adding more layers to model
model.add(Convolution2D(32, (3, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Dropout(0.25))
```

```python
# fully connected dense layers
model.add(Flatten())
model.add(Dense(128, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(10, activation='softmax'))
```

### Step 8. Compile model

```python
model.compile(loss='categorical_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])
```

### Step 8.5. Validation set

```python
# Create %20 validation set
X_val = X_train[:(len(X_train)/5)]
partial_X_train = X_train[(len(X_train)/5):]

Y_val = Y_train[:(len(X_train)/5)]
partial_Y_train = Y_train[(len(X_train)/5):]
```

### Step 9: Fit model on training data

```python
# normal
# history = model.fit(X_train, Y_train, batch_size=32, epochs=2, verbose=1)
```

```python
history = model.fit(partial_X_train,
                    partial_Y_train,
                    epochs=5,
                    batch_size=32,
                    verbose=1,
                    validation_data=(X_val, Y_val))
```

```python
print(history.history.keys())
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```

### Step 10: Evaluate model on test data

```python
score = model.evaluate(X_test, Y_test, verbose=0)
```
